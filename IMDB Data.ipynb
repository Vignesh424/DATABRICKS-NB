
## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
# File location and type
file_location = "/FileStore/tables/Game_of_Thrones.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
# Create a view or table

temp_table_name = "GOT"

df.createOrReplaceTempView(temp_table_name)
```

```
%sql

/* Query the created temp table in a SQL cell */

select * from GOT
```

```
from pyspark.sql import functions
from pyspark.sql.functions import *
```

```
#- How do you retrieve all columns from the dataset?

s1 = df.select("*").display(
)
```

```
#First 10 Rows of the Dataset
s2 = df.select("*").limit(10).display()
```

```
#How do you filter episodes from Season 1?
s3 = df.filter(col('Season')==1).select("*").display()
```

```
#Retrieve only the columns Season, Title of the Episode, and IMDb Rating.
s3 = df.select(col('Season'), col('Title of the Episode'),col('IMDb Rating')).display()
```

```
#Count the total number of episodes in the dataset.
s4  = print(df.count())
```

```
#Group episodes by Season and calculate the average IMDb rating per season.
s5 = df.filter(col('IMDB Rating').isNotNull()).groupBy("Season").agg(round(avg('IMDb Rating'),2).alias('IMDB')).display()
```

```
#Count the number of episodes directed by each director.
s6 = df.groupBy('directed by').count().sort(col('count').desc()).display()
```

```
#Find the highest-rated episode based on IMDb Ratings.
s7 = df.filter(col('IMDb Rating').isNotNull()).groupBy('Title of the Episode').agg(avg('IMDb Rating').alias('IMDB')).sort(col('IMDB').desc()).limit(1).display()
```

```
df = df.withColumn("Runtime Category", 
                   when(col("Running Time (Minutes)") < 50, "Short")
                   .when((col("Running Time (Minutes)") >= 50) & (col("Running Time (Minutes)") <= 60), "Medium")
                   .otherwise("Long"))

# Display the categorized episodes in Databricks
display(df.select("Title of the Episode", "Running Time (Minutes)", "Runtime Category"))
```

```
#IMDB Ratings
s9 = df.withColumn("IMDb Rating Desc",
                   when(col("IMDb Rating") <= 6, "Low Rating")
                   .when((col("IMDb Rating") > 6) & (col("IMDb Rating") <= 8), "Medium Rating")
                   .otherwise("High Rating")).display()

```

```
#User Viewership

s10 = df.withColumn("US Viewership Ratings",
                    when(col('U.S. Viewers (Millions)')<5, "Low Rated")
                    .when((col("U.S. Viewers (Millions)")>=5) & (col("U.S. Viewers (Millions)")<10), "Popular")
                    .otherwise("Blockbuster")
                    ).display()
```
